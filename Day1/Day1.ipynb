```
for column in data[data.columns[4:]]:
  if(data[column].isin([0]).sum()/487)*100>70:
    data.drop(column,axis=1,inplace=True)
```

```
data.shape
```

```
data
```

```
df=data[:].groupby("Country/Region").mean()
```

```
df.reset_index("Country/Region",inplace=True)
```

```
df
```

```
df.plot(kind='bar',x=df.iloc[:10,0],y=df['3/18/20'])
```

```

plt.bar(df.iloc[:10,0],df["3/4/20"][:10], color='g')
plt.legend(fontsize=14)
plt.xlabel("countries")
plt.ylabel('cases')

plt.title('country wise graph')

plt.show()
```

```
plt.barh(df.iloc[:10,0],df["3/4/20"][:10], color='g')
plt.legend(fontsize=14)
plt.xlabel("cases")
plt.ylabel('countries')

plt.title('country wise graph')

plt.show()
```

```
x=[date for date in df[df.columns[3:]]]
plt.bar(x,df.iloc[0,3:], color='g')
plt.legend(fontsize=14)
plt.xlabel("date")
plt.ylabel('cases')

plt.title('date wise graph of afganistan')

plt.show()
```
